# Exploratory Data Analysis (EDA) on the Titanic Dataset

**Objective:**
Analyze the Titanic dataset using Python to identify trends, patterns, relationships, and insights through statistical analysis and data visualization.

**Tools Used:**
- Python
- Pandas
- Matplotlib
- Seaborn



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside the notebook
%matplotlib inline

# Set plot style
sns.set_style("whitegrid")
```

```python
# Load the Titanic dataset
df = pd.read_csv("train.csv")

# Display the first 5 rows
df.head()
```

```python
# Dataset dimensions
print("Dataset Shape:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Dataset information
print("\nDataset Information:")
df.info()
```

**Observation**





*   The Titanic dataset contains passenger details such as age, gender, ticket class, fare, and survival status.  
*  Some columns contain missing values that will be explored later.


*   The dataset includes both numerical and categorical features for analysis.







```python
# Statistical summary of numerical columns
df.describe()
```

**Observation**


*   The statistical summary provides an overview of numerical features such as Age and Fare.
*  There is noticeable variation in passenger ages and ticket fares.



```python
# Check missing values
df.isnull().sum()
```

**Observation**





*   Missing values are present in Age, Cabin, and Embarked
*   Cabin has the highest number of missing values.




```python
print("Passenger Class")
print(df["Pclass"].value_counts())

print("\nGender")
print(df["Sex"].value_counts())

print("\nSurvival")
print(df["Survived"].value_counts())
```

**Observation**


*   Most passengers traveled in Third Class.
*   Male passengers outnumber female passengers.


*   More passengers did not survive than survived.






```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

**Observation**


*   More passengers did not survive than survived, indicating the disaster had a high mortality rate.




```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

**Observation**


* The dataset contains more male passengers than female passengers.



```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

**Observation**


*   Most passengers were between 20 and 40 years of age, with fewer elderly passengers.




```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title("Age Boxplot")
plt.show()
```

**Observation**
   
*   The age distribution contains several outliers, representing unusually old or young passengers.



```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```


**Observation**

*   Female passengers had a significantly higher survival rate than male passengers.



```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

**Observation**


*   First-class passengers had the highest survival rate, while third-class passengers had the lowest.



```python
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

**Observation**


*   Most passengers paid lower fares, while only a few passengers paid very high ticket prices.




```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='Blues', fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```

**Observation**


*   Passenger class and fare show relationships with survival, while some variables have only weak correlations.



```python
sns.pairplot(df[['Survived','Age','Fare','Pclass']].dropna(), hue='Survived')
plt.show()
```

**Observation**


*   The pairplot highlights patterns between age, fare, passenger class, and survival status.




```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

**Observation**

*   Higher fares are generally associated with first-class passengers, who also had better survival rates.



**Key Findings**


*  The dataset contains 891 passengers and 12 features.
*  Missing values were found in the Age, Cabin, and Embarked columns.

*  More passengers did not survive than survived.
*  Female passengers had a significantly higher survival rate than males.


*   First-class passengers had the highest survival rate.
*  Most passengers were between 20–40 years old.


*  Most passengers paid lower ticket fares, with a few paying exceptionally high fares.
*  Passenger class and fare influenced survival outcomes.


*   Exploratory Data Analysis helped identify important trends, patterns, and relationships within the dataset.
